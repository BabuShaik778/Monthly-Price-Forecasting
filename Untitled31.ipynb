```python
# -------------------------------------------
# Price Forecasting Project - Full Notebook
# -------------------------------------------


# -------------------------------
# Step 1: Import Libraries
# -------------------------------
import pandas as pd
import numpy as np
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# -------------------------------
# Step 2: Load Dataset
# -------------------------------
# Replace 'your_file.csv' with your uploaded CSV path
df = pd.read_csv('/content/price_data.csv')

# Inspect the dataset
print(df.head())
print(df.info())
print(df.describe())

# -------------------------------
# Step 3: Data Cleaning
# -------------------------------
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Sort by date
df.sort_values('date', inplace=True)

# Rename columns for Prophet
df.rename(columns={'date': 'ds', 'avg_monthly_price': 'y'}, inplace=True)

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Fill missing values (forward fill)
df['y'] = df['y'].fillna(method='ffill')

# Remove outliers using IQR
Q1 = df['y'].quantile(0.25)
Q3 = df['y'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['y'] >= Q1 - 1.5*IQR) & (df['y'] <= Q3 + 1.5*IQR)]

print(df.head())

# -------------------------------
# Step 4: Train Prophet Model
# -------------------------------
model = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
model.fit(df)

# -------------------------------
# Step 5: Forecast Next 12 Months
# -------------------------------
future = model.make_future_dataframe(periods=12, freq='M')
forecast = model.predict(future)

# Next 12 months forecast
forecast_next = forecast[['ds', 'yhat']].tail(12)
print("Forecast for next 12 months:")
print(forecast_next)

# -------------------------------
# Step 6: Evaluate Model
# -------------------------------
# Use last 12 months of historical data for evaluation
last_12_actual = df.tail(12)
last_12_forecast = forecast[['ds', 'yhat']].tail(12)

mae = mean_absolute_error(last_12_actual['y'], last_12_forecast['yhat'])
rmse = np.sqrt(mean_squared_error(last_12_actual['y'], last_12_forecast['yhat']))

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")

# -------------------------------
# Step 7: Plot Forecast
# -------------------------------
plt.figure(figsize=(12,6))
plt.plot(df['ds'], df['y'], label='Historical Price')
plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('Average Monthly Price')
plt.title('Price Forecast for Next 12 Months')
plt.legend()
plt.show()

# Prophet built-in plot
model.plot(forecast)
plt.show()

# Optional: Trend and components plot
model.plot_components(forecast)
plt.show()

# -------------------------------
# Step 8: Business Insights
# -------------------------------
# Example: Suggest actions based on forecast trend
forecast_next['trend'] = np.where(forecast_next['yhat'].diff() > 0, 'Increase', 'Decrease')
print("Next 12 months trend and suggested actions:")
print(forecast_next[['ds', 'yhat', 'trend']])
```

```python
# app.py
import streamlit as st
import pandas as pd
import numpy as np
import pickle
from prophet import Prophet
import matplotlib.pyplot as plt

st.set_page_config(page_title="Price Forecasting", layout="centered")

st.title(" Price Forecasting App")

# -------------------
# Step 1: Upload CSV
# -------------------
uploaded_file = st.file_uploader("Upload CSV file with columns: 'date', 'avg_monthly_price'", type=['csv'])

if uploaded_file:
    df = pd.read_csv(uploaded_file)
    st.subheader("Raw Data")
    st.dataframe(df.head())

    # Preprocess
    df['date'] = pd.to_datetime(df['date'])
    df.sort_values('date', inplace=True)
    df.rename(columns={'date': 'ds', 'avg_monthly_price': 'y'}, inplace=True)
    df['y'] = df['y'].fillna(method='ffill')

    # -------------------
    # Step 2: Load Model
    # -------------------
    try:
        with open('prophet_model.pkl', 'rb') as f:
            model = pickle.load(f)
        st.success("Model loaded successfully!")
    except FileNotFoundError:
        st.error("Pickle model not found! Train and save the model first.")
        st.stop()

    # -------------------
    # Step 3: Forecast
    # -------------------
    periods = st.slider("Select number of months to forecast:", min_value=1, max_value=24, value=12)
    future = model.make_future_dataframe(periods=periods, freq='M')
    forecast = model.predict(future)

    st.subheader("Forecasted Prices")
    st.dataframe(forecast[['ds', 'yhat']].tail(periods))

    # Trend suggestion
    forecast_next = forecast[['ds', 'yhat']].tail(periods)
    forecast_next['trend'] = np.where(forecast_next['yhat'].diff() > 0, 'Increase', 'Decrease')
    st.subheader("Forecast Trend")
    st.dataframe(forecast_next)

    # -------------------
    # Step 4: Plot Forecast
    # -------------------
    st.subheader("Forecast Plot")
    plt.figure(figsize=(12,6))
    plt.plot(df['ds'], df['y'], label='Historical Price')
    plt.plot(forecast['ds'], forecast['yhat'], label='Forecast', color='red')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    st.pyplot(plt)
```
